```python
# Create a student database having Name, Roll, Course, Marks, Attendance. Perform the
# following operations:
# ● Find Students scoring above 80 in Math.
# ● Class Topper using Average Marks.
# ● Students enrolled in both Course A and Course B
# ● Find Students with lowest Attendance below 40.
# ● Students who Took All 3 Subjects.

db = [
    {
        "name": "Krish",
        "roll": 1,
        "course": ["A", "B"],
        "marks": {
            "Math": 59,
            "Phy": 68,
            "ComA": 98 
        },
        "attendance": 87
    },
    {
        "name": "Souvik",
        "roll": 2,
        "course": ["A", "B", "C"],
        "marks": {
            "Math": 72,
            "Phy": 50,
            "ComA": 82 
        },
        "attendance": 98
    },
    {
        "name": "Sapt",
        "roll": 3,
        "course": ["A", "C"],
        "marks": {
            "Math": 89,
            "Phy": 90,
            "ComA": 80 
        },
        "attendance": 91
    },
    {
        "name": "Harry",
        "roll": 4,
        "course": ["B", "C"],
        "marks": {
            "Math": 83,
            "Phy": 77,
            "ComA": 99 
        },
        "attendance": 32
    }
]

print("Students how score above 80 in Maths:")
for record in db:
    if record["marks"]["Math"] > 80:
        print(record["name"])

print("Class topper:")
topper = max(db, key=lambda x: sum(x["marks"].values())/len(x["marks"]))
print(topper["name"])

print("Students enrolled in both Course A and Course B: ")
for record in db:
    if "A" in record["course"] and "B" in record["course"]:
        print(record["name"])

print("Find Students with lowest Attendance below 40: ")
for record in db:
    if record["attendance"] < 40:
        print(record["name"])

print("Students who Took All 3 Subjects: ")
for record in db:
    if len(record["course"]) >= 3:
        print(record["name"])
```

```python
from collections import deque

def bfs(adj, start, n):
    visited = [0 for _ in range(n+1)]
    queue = deque()
    queue.append(start)
    visited[start] = 1

    print("BFS Traversal: ", end=" ")
    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbor in adj[node]:
            if visited[neighbor] == 0:
                visited[neighbor] = 1
                queue.append(neighbor)
    print()

def main():
    n = int(input("Enter number of node: "))
    e = int(input("Enter number of edge: "))

    adj = [[] for _ in range(n+1)]

    print("Enter the edge (u, v): ")
    for _ in range(e):
        u, v = map(int, input().split())
        adj[u].append(v)
        adj[v].append(u)

    start = int(input("Enter the start node: "))

    bfs(adj, start, n)

main()
```

```python
# Write a python code for implementing DFS.


def dfs_util(adj, node, visited):
    visited[node] = 1
    print(node, end=" ")

    for neighbor in adj[node]:
        if visited[neighbor] == 0:
            dfs_util(adj, neighbor, visited)

def dfs(adj, start, n):
    visited = [0 for _ in range(n+1)]
    print("DFS Traversal: ", end=" ")
    dfs_util(adj, start, visited)
    print()

def main():
    n = int(input("Enter number of node: "))
    e = int(input("Enter number of edge: "))

    adj = [[] for _ in range(n+1)]

    print("Enter Edge (u, v): ")
    for _ in range(e):
        u, v = map(int, input().split())
        adj[u].append(v)
        adj[v].append(u)

    start = int(input("Enter the start node: "))

    dfs(adj, start, n)

main()
```

```python
# Water juge Problem

from collections import deque

def print_step(path):
    print("Step to reach target: ")
    for step in path:
        print(step)

def water_jug_bfs(jug1, jug2, target):
    visited = set()
    queue = deque()
    queue.append(((0, 0), []))
    while queue:
        (a, b), path = queue.popleft()
        if a == target or b == target:
            path.append((a, b))
            print_step(path)
            return True
        if (a, b) not in visited:
            visited.add((a, b))
            moves = []
            moves.append((jug1, b))
            moves.append((a, jug2))
            moves.append((0, b))
            moves.append((a, 0))
            pour = min(a, jug2 - b)
            moves.append((a - pour, b + pour))
            pour = min(b, jug1 - a)
            moves.append((a + pour, b - pour))
            for new_state in moves:
                if new_state not in visited:
                    queue.append((new_state, path+[(a, b)]))
    print("No soulition found!")
    return False

def main():
    jug1 = int(input("Enter the capacity of jug1: "))
    jug2 = int(input("Enter the capacity of jug2: "))
    target = int(input("Enter the target: "))
    water_jug_bfs(jug1, jug2, target)

main()
```

```python
# Simple Tic Tac Toe (Human vs Human)

def print_board(board):
    for row in board:
        print(" | ".join(row))
    print("-" * 9)

def generate_winning_position(n=3):
    positions = []
    for r in range(n):
        positions.append([(r, c) for c in range(n)])
    for c in range(n):
        positions.append([(r, c) for r in range(n)])

    positions.append([(i, i) for i in range(n)])
    positions.append([(i, n-1-i) for i in range(n)])
    return positions

def check_winnier(board, symbol):
    for positions in generate_winning_position():
        if all(board[r][c] == symbol for r, c in positions):
            return positions
    return None 

def is_draw(board):
    return all(cell != "." for row in board for cell in row)

def tictactoi():
    board = [["." for _ in range(3)] for _ in range(3)]
    players = [("Player 1", "O"), ("Player 2", "X")]
    turn = 0

    while True:
        print_board(board)
        player, symbol = players[turn]
        print(f"{player}'s turn ({symbol}) ")

        try:
            row = int(input("Enter row index (0-2): "))
            col = int(input("Enter col index (0-2): "))
        except ValueError:
            print("Invalid Input, Please try 0, 1, or 2..")
            continue

        if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == ".":
            board[row][col] = symbol
        else:
            print("Invalid Move! Try again")
            continue

        if check_winnier(board, symbol):
            print_board(board)
            print(f"{player} wins!")
            break

        if is_draw(board):
            print_board(board)
            print("It is draw!")
            break


        turn = 1 - turn
```

```python
# Simple 8 Puzzle Solver using BFS

from collections import deque

def print_board(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print("-" * 9)

def get_neighbors(state):
    neighbors = []
    idx = state.index(0)  # empty tile
    row, col = divmod(idx, 3)
    moves = [(-1,0),(1,0),(0,-1),(0,1)]  # up, down, left, right
    for dr, dc in moves:
        r, c = row+dr, col+dc
        if 0 <= r < 3 and 0 <= c < 3:
            new_idx = r*3+c
            new_state = list(state)
            new_state[idx], new_state[new_idx] = new_state[new_idx], new_state[idx]
            neighbors.append(tuple(new_state))
    return neighbors

def bfs(start, goal):
    queue = deque([(start, [])])
    visited = {start}
    while queue:
        state, path = queue.popleft()
        if state == goal:
            return path + [state]
        for neighbor in get_neighbors(state):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path+[state]))
    return None

def main():
    # User input
    start = tuple(map(int, input("Enter initial state (9 numbers, use 0 for empty): ").split()))
    goal = tuple(map(int, input("Enter target state (9 numbers, use 0 for empty): ").split()))

    print("\nSolving...\n")
    solution = bfs(start, goal)

    if solution:
        for step in solution:
            print_board(step)
        print("Solved in", len(solution)-1, "moves!")
    else:
        print("No solution found.")

main()
```
