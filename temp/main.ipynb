```python
# code by Krish
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 3, 5, 7, 9]
print("Linear Search:", linear_search(arr, 7))
print("Binary Search:", binary_search(arr, 7))
```

```python
from collections import deque
# code by Krish
def print_state(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()

def get_neighbors(state):
    neighbors = []
    idx = state.index(0)
    row, col = divmod(idx, 3)

    moves = [(-1,0), (1,0), (0,-1), (0,1)]
    for dr, dc in moves:
        new_row, new_col = row + dr, col + dc
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_idx = new_row * 3 + new_col
            new_state = list(state)
            new_state[idx], new_state[new_idx] = new_state[new_idx], new_state[idx]
            neighbors.append(tuple(new_state))
    return neighbors

def bfs(start, goal):
    queue = deque([(start, [])])
    visited = set([start])

    while queue:
        state, path = queue.popleft()
        if state == goal:
            return path + [state]

        for neighbor in get_neighbors(state):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [state]))
    return None

print("Enter the start state (use 0 for empty tile):")
start_state = tuple(map(int, input().split()))

print("Enter the goal state (use 0 for empty tile):")
goal_state = tuple(map(int, input().split()))

solution = bfs(start_state, goal_state)

if solution:
    print("Solution found in", len(solution)-1, "moves:")
    for step in solution:
        print_state(step)
else:
    print("No solution exists.")
```

```python
def ana_para(paragraph):
    vowels = "aeiou"+"AEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0
    space_count = 0

    for char in paragraph:
        if char.isdigit():
            digit_count += 1
        elif char.isspace():
            space_count += 1
        elif char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    print("Vowels:", vowel_count)
    print("Consonants:", consonant_count)
    print("Digits:", digit_count)
    print("Spaces:", space_count)
    
    words = paragraph.split()

    longest_word = max(words, key=len)
    print("Longest word:", longest_word)

    shortest_word = min(words, key=len)
    print("Shortest word:", shortest_word)

    reversed_sentence = paragraph[::-1]
    print("Reversed sentence:", reversed_sentence)

def main():
    paragraph = input("Enter a paragraph: ")
    ana_para(paragraph)

main()
```

```python
def calculate_grade(avg):
    if avg >= 90:
        return "O"
    elif avg >= 80:
        return "E"
    elif avg >= 70:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 50:
        return "C"
    elif avg >= 40:
        return "D"
    elif avg < 40:
        return "F"
    else:
        return "I" #incomplete according to makaut

N = int(input("Enter the number of students: "))

students = []  

for i in range(N):
    print(f"\nEnter marks for Student {i+1}:")
    name = input("Name: ")
    marks = []
    subject = ["DSA", "AI", "Cloud"]
    for j in range(3):
        m = int(input(f"{subject[j]} Subject marks: "))
        marks.append(m)

    total = sum(marks)
    avg = total / 3
    grade = calculate_grade(avg)

    students.append({
        "name": name,
        "marks": marks,
        "total": total,
        "average": avg,
        "grade": grade
    })

print("\n\t---------------- Student Results ----------------")
for s in students:
    print(f"Name: {s['name']}, Total: {s['total']}, Average: {s['average']:.2f}, Grade: {s['grade']}")

topper = max(students, key=lambda x: x["total"])
print(f"\nTopper: {topper['name']} with {topper['total']} marks")

class_avg = sum(s["average"] for s in students) / N
print(f"Class Average: {class_avg:.2f}")
```

```python
#bfs
from collections import deque

def build_adj_matrix():
    n = int(input("Enter the number of nodes: "))
    print(f"Number of nodes: {n}")

    e = int(input("Enter the number of edge: "))
    print(f"Number of edge: {e}")

    adj_matrix = [[] for _ in range(n+1)]

    print("Enter the edgs connection as formate of: u v")
    print("Where u = from node, v = to node")
    for _ in range(e):
        u, v = map(int, input().split())
        adj_matrix[u].append(v)
        adj_matrix[v].append(u)
    return adj_matrix

def bfs(adj_matrix, start):
    n = len(adj_matrix)
    visited = [0] * (n+1)
    queue = deque()
    queue.append(start)
    visited[start] = 1

    print("BFS Traversal: ")
    while queue:
        node = queue.popleft()
        print(f"{node} ", end="")

        for neighbor in adj_matrix[node]:
            if visited[neighbor] == 0:
                visited[neighbor] = 1
                queue.append(neighbor)
    print()

def main():
    adj_matrix = build_adj_matrix()
    start = int(input("Enter the start node: "))
    print(f"Start node is: {start}")

    bfs(adj_matrix, start)

main()
```

```python
#dfs
def build_adj_matrix():
    n = int(input("Enter the number of nodes: "))
    print(f"Number of nodes: {n}")
    e = int(input("Enter the number of edge: "))
    print(f"Number of edge: {e}")

    adj_matrix = [[] for _ in range(n+1)]

    print("Enter the edge connection as formate of: u v")
    print("Where u = from node, v = to node")
    for _ in range(e):
        u, v = map(int, input().split())
        adj_matrix[u].append(v)
        adj_matrix[v].append(u)
    return adj_matrix

def dfs_util(adj_matrix, node, visited):
    visited[node] = 1
    print(f"{node} ", end="")

    for neighbor in adj_matrix[node]:
        if visited[neighbor] == 0:
            dfs_util(adj_matrix, neighbor, visited)

def dfs(adj_matrix, start):
    n = len(adj_matrix)
    visited = [0] * (n+1)
    print("DFS Traversal: ", end="")
    dfs_util(adj_matrix, start, visited)
    print()

def main():
    adj_matrix = build_adj_matrix()

    print("Adj_Matrix is: ", adj_matrix)

    start = int(input("Enter the start node: "))
    print(f"Start node is: {start}")

    dfs(adj_matrix, start)

main()
```

```python
# Water jug
def print_step(path):
    print("Steps to reach the target: ")
    for step in path:
        print(step)
    print()

def water_jug(jug1, jug2, target):
    visited = set()
    queue = deque()

    queue.append(((0,0), []))

    while queue:
        (a, b), path = queue.popleft()
        if a == target and b == target:
            path.append((a, b))
            print_step(path)
            return True
        if (a, b) in visited:
            continue
        visited.add((a, b))
        moves = []
        moves.append((jug1, b))
        moves.append((a, jug2))
        moves.append((0, b))
        moves.append((a, 0))
        pour = min(a, jug2 - b)
        moves.append((a - pour, b + pour))
        pour = min(b, jug1 - a)
        moves.append((a + pour, b - pour))
        for state in moves:
            if state not in visited:
                queue.append((state, path + [(a, b)]))
    print("No solution found!")
    return False

def main():
    jug1 = int(input("Enter the capacity of jug1: "))
    jug2 = int(input("Enter the capcacity of jug2: "))
    target = int(input("Enter the target amount: "))

    water_jug(jug1, jug2, target)

main()
```

```python
import random
# code by Krish
def print_board(board):
    for row in board:
        print(" | ".join(row))
    print("-" * 9)

def generate_winning_positions(n=3):
    positions = []
    for r in range(n):
        positions.append([(r, c) for c in range(n)])
    for c in range(n):
        positions.append([(r, c) for r in range(n)])
    positions.append([(i, i) for i in range(n)])
    positions.append([(i, n - 1 - i) for i in range(n)])
    return positions

def check_winner(board, symbol):
    winning_positions = generate_winning_positions()
    for positions in winning_positions:
        if all(board[r][c] == symbol for r, c in positions):
            return positions
    return None

def is_draw(board):
    return all(cell != '.' for row in board for cell in row)

def ai_move(board, ai_symbol, human_symbol):
    for r in range(3):
        for c in range(3):
            if board[r][c] == '.':
                board[r][c] = ai_symbol
                if check_winner(board, ai_symbol):
                    return r, c
                board[r][c] = '.'

    for r in range(3):
        for c in range(3):
            if board[r][c] == '.':
                board[r][c] = human_symbol
                if check_winner(board, human_symbol):
                    board[r][c] = ai_symbol
                    return r, c
                board[r][c] = '.'

    if board[1][1] == '.':
        board[1][1] = ai_symbol
        return 1, 1

    for r, c in [(0,0),(0,2),(2,0),(2,2)]:
        if board[r][c] == '.':
            board[r][c] = ai_symbol
            return r, c

    free_cells = [(r,c) for r in range(3) for c in range(3) if board[r][c] == '.']
    r, c = random.choice(free_cells)
    board[r][c] = ai_symbol
    return r, c

def tic_tac_toe():
    board = [['.' for _ in range(3)] for _ in range(3)]

    mode = input("\tPlay as \n-----------------------\nPlayer 1 (vs AI) or \nPlayer 2 (vs Human)? \nEnter 1 or 2: ")

    if mode == '1':
        players = [('Human', 'O'), ('AI', 'X')]
    else:
        players = [('Player 1', 'O'), ('Player 2', 'X')]

    turn = 0

    while True:
        print_board(board)
        player, symbol = players[turn]
        print(f"{player}'s turn ({symbol})")

        if player == "AI":
            row, col = ai_move(board, symbol, players[0][1])
            print(f"AI chooses: {row}, {col}")
        else:
            try:
                row = int(input("Enter row (0-2): "))
                col = int(input("Enter col (0-2): "))
            except ValueError:
                print("Invalid input! Please enter numbers 0, 1, or 2.")
                continue

            if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == '.':
                board[row][col] = symbol
            else:
                print("Invalid move! Try again.")
                continue

        winning_positions = check_winner(board, symbol)
        if winning_positions:
            print_board(board)
            print(f"{player} wins! Winning positions: {winning_positions}")
            break

        if is_draw(board):
            print_board(board)
            print("It's a draw!")
            break

        turn = 1 - turn

if __name__ == "__main__":
    tic_tac_toe()
```
